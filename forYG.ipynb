Use panda to sort out the excel form

Step 1: Import pandas and read the file from excel file

```python
import pandas as pd

fileName = 'data1.xlsx'
dataTable = pd.read_excel(fileName)
dataTable
```

Step 2: manipulate the table

Group them by isin_nm and put each group into one row

```python
gp = dataTable.groupby('isin_nm')
# make a new data frame to put the data in
# sort group index by size, put bigger group into the new frame first
isinSorted = gp.size().sort_values(ascending=False).index
maxColumn = gp.size().values.max()*3 + 1
dataTableNew = pd.DataFrame(columns=['isin_nm']+['rate_eff_dt', 'cpn_rate', 'rate_end_dt']*gp.size().values.max())
for i, isin in enumerate(isinSorted):
    recordList = [isin]  # put the isin into the list first
    size = gp.size()[isin]  # get the size
    # loop throught the group, put the record into the list
    for j in gp.get_group(isin).index:
        recordList += gp.get_group(isin).loc[j].tolist()[1:]
    nMissingColumn = maxColumn - len(recordList)
    if nMissingColumn > 0:
        recordList += [None] * nMissingColumn
    print(len(recordList))
    dataTableNew.loc[i] = recordList
dataTableNew        
```

Step 3 save the new data frame.



```python
dataTableNew.to_excel('newdata.xlsx', index=False)
```
